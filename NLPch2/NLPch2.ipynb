### 모든 책에 대해 글자수/단어 , 단어수/문장수 , 단어수/단어집합 계산

````python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print (int(num_chars/num_words), int(num_words/num_sents), int(num_words/num_vocab),fileid)
````

# Plotting and Tabulating Distributions

ConditionalFreqDist 은 유용한 plotting과 tabulaiting(표로만들기) 함수를 제공한다

### Inaugural Address Corpus

미국 대통령들의 취임식 연설문 데이터

밑의 예제는 america와 citizen 이라는 단어가 연설문들에 몇번씩 나왔는지 ploting 하는 코드

````python
nltk.download('inaugural')
````

````python
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
(target, fileid[:4])
for fileid in inaugural.fileids()
for w in inaugural.words(fileid)
for target in ['america', 'citizen']
if w.lower().startswith(target))
cfd.plot()
````

# Generating Random Text with Bigrams

```python
>>> sent = ['In', 'the', 'beginning', 'God', 'created', 'the', 'heaven',
... 'and', 'the', 'earth', '.']
>>> nltk.bigrams(sent)
[('In', 'the'), ('the', 'beginning'), ('beginning', 'God'), ('God', 'created'),
('created', 'the'), ('the', 'heaven'), ('heaven', 'and'), ('and', 'the'),
('the', 'earth'), ('earth', '.')]
```
 
이전 장에서 살펴봤떤 biagrams 을 이용하여 임의의 문장을 생성가능하다.

example 2-1 에서 각각의 단어를 condition으로 보고 각각 단어에 따른 Conditional Frequency Distributions 을 구한다. generate_model function은 주어진 단어의 cfdist가 최대인 단어를 return 하는 간단한 반복문을 수행한다.

````python
def generate_model(cfdist, word, num=15):
    for i in range(num):
        print (word)
        word = cfdist[word].max()
text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
````

````python
print (cfd['living'])
````

````python
generate_model(cfd, 'living')
````
